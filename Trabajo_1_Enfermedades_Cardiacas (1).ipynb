```r
library(tibble)
library(dplyr)
## Cargamos las librerias para realizar la trasnformacion e imputacion de los datos
datos_transformado <- datos_csv %>% # Convertiremos las variables categoricas a factor incluyendo algunas con valores booleanos
  remove_rownames() %>%
  column_to_rownames(var = "id") %>%
  mutate(
    across(
      c(sex, cp, restecg, slope, fbs, exang),
      as.factor
    ),
    # Variables de conteo a enteros
    age = as.integer(age),
    # Variables continuas como tipo numeric , en vista que no queremos perder decimales para su calculo
    trestbps = as.numeric(trestbps),
    chol = as.numeric(chol),
    thalch = as.numeric(thalch),
    oldpeak = as.numeric(oldpeak)
  ) %>%

  select(-thal, -ca, -dataset) ## Descartamos esta variables por la que mas del 40% de sus datos son faltantes y dataset por que no aporta nada al analisis
```

```r
summary(datos_transformado) ## Revision de los datos transformados
```

```r
library(naniar) ## Revision de la base de datos transformada
gg_miss_var(datos_transformado)
```

```r
library(mice) ## Se realiza la imputacion con esta libreria , seleccionando cada tipo de metodo segun la variable , aidicionalmente se incluye las variables que estan completas
              ## sin embargo, la funcion a la señalar con "" , no las imputa
imputacion_vars <- c("trestbps", "chol", "thalch", "oldpeak", "exang", "fbs","slope","age","sex","num","cp","restecg")
metodos <- c("pmm","pmm","pmm","norm","logreg","logreg","polyreg","","","","","")
imputed <- mice(datos_transformado[, imputacion_vars], m = 1, method = metodos, seed = 123)
datos_imp <- complete(imputed)
```

```r
library(naniar)
gg_miss_var (datos_imp)
```

#### Podemos ver que incluso al imputar con una libreria robusta aun no es posible totalmente eliminar los datos faltantes, por eso estas variables no pueden ser usadas para el anailisis, sin embargo las variables continuas fueron imputadas con exito

```r
library(dplyr)
## Transformamos la variable num a binaria para poder realizar una compracion entre las personas que tienen la enfermdad y las que no tienen sin importar su etapa de gravedad
datos_imp <- datos_imp %>%
  mutate(enfermedad = ifelse(num == 0, "No", "Sí"))
```

```r
localizacion <- datos_imp %>%
  group_by(enfermedad) %>%
  summarise(
    across(
      .cols = c(age, trestbps, chol, thalch, oldpeak),
      .fns = list(
        Min = ~min(.x, na.rm = TRUE),
        Max = ~max(.x, na.rm = TRUE),
        Media = ~mean(.x, na.rm = TRUE),
        Mediana = ~median(.x, na.rm = TRUE),
        Q1 = ~quantile(.x, 0.25, na.rm = TRUE),
        Q3 = ~quantile(.x, 0.75, na.rm = TRUE)
      )
    )
  )
```

## Medidas de dispersion
$ rango= maximo - minimo


* Varianza / Desviacion estandar

  \sigma =\sqrt{\frac{\sum _{i=1}^{N}(x_{i}-\mu )^{2}}{N}}\

$ Desviacion estandar = \sigma = \sqrt{\sigma^2}


* $ Coeficiente de variacion (CV)   =   \frac{s}{|\bar{x}|}*100\%$

```r
dispersion <- datos_imp %>%
  group_by(enfermedad) %>%
  summarise(
    across(
      .cols = c(age, trestbps, chol, thalch, oldpeak),
      .fns = list(
        Min = ~min(.x, na.rm = TRUE),
        Max = ~max(.x, na.rm = TRUE),
        Media = ~mean(.x, na.rm = TRUE),
        Mediana = ~median(.x, na.rm = TRUE),
        Rango = ~ (max(.x, na.rm = TRUE) - min(.x, na.rm = TRUE)),
        Varianza = ~var(.x, na.rm = TRUE),
        SD = ~sd(.x, na.rm = TRUE),
        CV = ~ (sd(.x, na.rm = TRUE) / mean(.x, na.rm = TRUE)) * 100,
        IQR = ~IQR(.x, na.rm = TRUE)
      )
    )
  )
```

```r
library(formattable) ## Para una mejor visualizacion dividiremos en varias secciones los calculos
dispersion %>%
  select(enfermedad, age_Min, age_Max, age_Rango, age_Mediana, age_Varianza, age_SD, age_CV, age_IQR, trestbps_Min, trestbps_Max, trestbps_Rango, trestbps_Mediana,trestbps_Varianza, trestbps_SD, trestbps_CV, trestbps_IQR) %>%
  formattable()
```

```r
library(formattable) ## Para una mejor visualizacion dividiremos en varias secciones los calculos
dispersion %>%
  select(enfermedad, chol_Min, chol_Max, chol_Rango, chol_Mediana, chol_Varianza, chol_SD, chol_CV, chol_IQR, thalch_Min, thalch_Max, thalch_Rango, thalch_Mediana,thalch_Varianza, thalch_SD, thalch_CV, thalch_IQR) %>%
  formattable()
```

```r
library(formattable) ## Para una mejor visualizacion dividiremos en varias secciones los calculos
dispersion %>%
  select(enfermedad, oldpeak_Min, oldpeak_Max, oldpeak_Rango, oldpeak_Mediana, oldpeak_Varianza, oldpeak_SD, oldpeak_CV, oldpeak_IQR) %>%
  formattable()
```

##### Despues de realizar los calculos de las medidas de dispersion nos podemos notar las gran dispersion de que existe en los valores del colesterol serico debido al alto porcentaje de coeficiente de varianza.
